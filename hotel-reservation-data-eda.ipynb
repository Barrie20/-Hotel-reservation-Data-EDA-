## 3.1 Make "feature"- counts bar graph

<b> it is useful to analysis simple relatioship between feature and its counts.

```python
def make_bar_graph(feature):
    if type(df[str(feature)].value_counts().index[0]) == str:
        x = df[str(feature)].value_counts().index
        y = df[str(feature)].value_counts().values
    else:
        x = df[str(feature)].value_counts().sort_index().index
        y = df[str(feature)].value_counts().sort_index().values
    plt.figure(figsize=(20,8))
    for i in range(len(x)):
        height = y[i]
        plt.text(x[i], height + 0.25, '%.1f' %height, ha='center', va='bottom', size = 8)
    plt.title(str(feature))
    if len(x) <= 2:
        return plt.bar(x,y,color='#e35f62')
    elif len(x) > 2:
        return plt.plot(x,y,color='#e35f62')
```

## 3.2 Make Each Featurs' Graph

```python
for i in df.columns[1:]:
    make_bar_graph(i)
```

```python
for i in [2017,2018]: # There is only 2years(2017,2018) data in dataset
    x = df[df["arrival_year"]==i]["arrival_month"].value_counts().sort_index().index
    y = df[df["arrival_year"]==i]["arrival_month"].value_counts().sort_index().values
    plt.figure(figsize=(15,8))
    plt.subplot(2,1,i-2016)
    plt.xlabel("Month")
    plt.ylabel("Number of visitors")
    plt.title(i)
    for b in range(len(x)):
        height = y[b]
        plt.text(x[b], height + 0.25, '%.1f' %height, ha='center', va='bottom', size = 12)
    plt.bar(x,y,color='#e35f62')
```

### Make the defnition to visualize "Number of '<mark>Feature</mark>'" - "<mark>Date</mark>" relationship

```python
def date_numb_graph(feature):
    for year in [2017,2018]:
        plt.figure(figsize=(15,15))
        plt.subplot(2,1,year-2016)
        for i in range(df[df["arrival_year"]==year]["arrival_month"].value_counts().sort_index().index[0],13):
            a = df[(df["arrival_year"]==year)&(df["arrival_month"]==i)][str(feature)].value_counts().sort_index().plot(label=(i,"month"))
            x = df[(df["arrival_year"]==year)&(df["arrival_month"]==i)][str(feature)].value_counts().sort_index().index
            y = df[(df["arrival_year"]==year)&(df["arrival_month"]==i)][str(feature)].value_counts().sort_index().values
            for b in range(len(x)):
                height = y[b]
                plt.text(x[b], height + 0.25, '%.1f' %height, ha='center', va='bottom', size = 12)
        plt.legend()
        
        plt.xlabel(str(feature))
        plt.ylabel(str(feature) +" per month in "+ str(year))
        plt.title(str(feature) +" per month in "+ str(year)+ " - Number of Reservation")
```

## 5.2 market_segment_type

#### 5.2.1 Create New Market Segment Type Dataframe by each year 

```python
mst = pd.DataFrame()
mst.insert(0,"2017",df[df["arrival_year"]==2017]["market_segment_type"].value_counts().values)
mst.insert(0,"2018",df[df["arrival_year"]==2018]["market_segment_type"].value_counts().values[:4])
mst_index = ["Online","Offline","Corporate","Complementary"]
mst.drop
mst.insert(0,"type",mst_index)
```

```python
mst.set_index("type",inplace=True)
mst
```

#### 5.2.2 Visualization of Market Segment Types by Year

```python
mst.plot(kind='bar', rot=0,figsize=(20,8),color=['#e35f62',"blue"])
plt.title('market_segment_type', fontsize=20)
plt.legend(fontsize='12')
plt.show()
```

#### 5.2.3 Visualization Percentage of Market Segment Types by Year

#### From the graph above, it can be seen that the number of people booked online in 2018 is remarkably high. However, this result may be derived because the 2017 data present on the data is less than the 2018 data. So, I try to check the trend of people's reservation methods in more detail through the ratio of reservation methods.

```python
plt.figure(figsize=(15,15))
for i in [2017,2018]:
    ratio = mst[str(i)]
    plt.subplot(1,2,i-2016)
    plt.title(str(i))
    plt.pie(ratio, labels=mst.index, autopct='%.1f%%', counterclock=False,explode=(0.01,0.01,0.01,0.02))
```

#### It can be seen that within a year, the number of offline reservations decreased significantly and the number of visitors booking hotels online increased significantly. These results can prove that over time, people use the Internet to book hotels more frequently.

### 5.3.1 Mapping _ price

```python
df.loc[:, "avg_price_map"] = 0
```

```python
for i in range(len(df["avg_price_per_room"])):
    if len(str(df["avg_price_per_room"][i]).split('.')[0]) == 2:
        df["avg_price_map"][i] = int(str(df["avg_price_per_room"][i]).split('.')[0][:1])
    elif len(str(df["avg_price_per_room"][i]).split('.')[0]) == 3:
        df["avg_price_map"][i] = int(str(df["avg_price_per_room"][i]).split('.')[0][:2])
    else:
        df["avg_price_map"][i] = 0
```

```python
df["avg_price_map"].value_counts().sort_index().head()
```

### 5.3.2 Visualize Mapped Average Price Per Room

```python
x = df["avg_price_map"].value_counts().sort_index().index.tolist()
y = df["avg_price_map"].value_counts().sort_index().values.tolist()
plt.figure(figsize=(20,8))
for i in range(len(x)):
    height = y[i]
    plt.text(x[i], height + 0.25, '%.1f' %height, ha='center', va='bottom', size = 8)
plt.xlabel("Mapped Average payment (*10$)")
plt.ylabel("total guests' number")
plt.bar(x,y,color="#e35f62")
```

#### Most of the visitor paid 60~120$
